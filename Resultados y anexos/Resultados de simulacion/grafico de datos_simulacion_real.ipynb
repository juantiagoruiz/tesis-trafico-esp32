```python
#imports Matplotlib library and assigns shorthand 'plt'
import matplotlib.pyplot as plt
#imports Seaborn library and assigns shorthand 'sns'
import seaborn as sns
#imports Plotly library and assigns shorthand 'px'
#import plotly.express as px
import pandas as pd
import numpy as np
```

```python
def generate_bar_traveltime():
    
    Tiempo_viaje_1_L = 73.83
    Tiempo_viaje_2_L = 91.64
    Tiempo_viaje_3_L = 73.02

    Tiempo_viaje_1_M = 79.10
    Tiempo_viaje_2_M = 97.37
    Tiempo_viaje_3_M = 77.70

    Tiempo_viaje_1_H = 83.61
    Tiempo_viaje_2_H = 98.16
    Tiempo_viaje_3_H = 80.75

    algoritmos = ['ATF','ACAB','ACAB Realista']
    
    promedios_L = [Tiempo_viaje_2_L,Tiempo_viaje_3_L,Tiempo_viaje_1_L]
    promedios_M = [Tiempo_viaje_2_M,Tiempo_viaje_3_M,Tiempo_viaje_1_M]
    promedios_H = [Tiempo_viaje_2_H,Tiempo_viaje_3_H,Tiempo_viaje_1_H] 

    print(((Tiempo_viaje_3_H+Tiempo_viaje_1_H)/2)/Tiempo_viaje_2_H)
    x = np.arange(len(algoritmos))
    width = 0.2
    plt.figure(figsize=(10,6), dpi=600)
    plt.bar(x-width/2-width/2, promedios_L,width,label='Densidad trafico baja')
    plt.bar(x, promedios_M,width,label='Densidad trafico media')
    plt.bar(x+width/2+width/2, promedios_H,width,label='Densidad trafico alta')

    plt.xticks(x, algoritmos)
    plt.legend()
    plt.title('Promedio de tiempo de viaje')
    plt.ylabel('Duracin de viaje en segundos')
    plt.xlabel('Algoritmos')
    plt.ylim(0,130)
    plt.grid()
    plt.show()

generate_bar_traveltime()
    
```

```python
def generate_graphs(data_1, data_2, data_3, variable):

    # Load the dataset into a Pandas DataFrame
    df_1_light = pd.read_csv(data_1)
    variable = variable

    # Rename the column to remove an additional space
    #df = df.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})

    # Extract the date and close price columns
    step_history = df_1_light['step_history']
    variable_history = df_1_light[variable]

    # Create a line plot
    plt.plot(step_history/20, variable_history,label='AACA')
    if variable == 'emissions_history':
        plt.title('Emisiones por tiempo trfico liviano algoritmo 1')
        plt.ylabel('Emisiones de CO2 en g/s')
    plt.xlabel('Segundos de simulacin')

    df_2_light = pd.read_csv(data_2)

    # Rename the column to remove an additional space
    #df = df.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})

    # Extract the date and close price columns
    step_history = df_2_light['step_history']
    variable_history = df_2_light[variable]
    # Create a line plot
    plt.plot(step_history/20, variable_history,label='ACAB')
    if variable == 'emissions_history':
        plt.title('Emisiones por tiempo trfico liviano algoritmo 3')
        plt.ylabel('Emisiones de CO2 en g/s')
    plt.xlabel('Segundos de simulacin')


    df_3_light = pd.read_csv(data_3)

    # Extract the date and close price columns
    step_history = df_3_light['step_history']
    variable_history = df_3_light[variable]
    # Create a line plot
    plt.plot(step_history/20, variable_history,label='ATF')
    if variable == 'emissions_history':
        plt.title('Emisiones por tiempo trfico liviano algoritmo 3')
        plt.ylabel('Emisiones de CO2 en g/s')
    plt.xlabel('Segundos de simulacin')
    plt.legend()
    plt.show

    print(df_1_light[variable].mean())
    print(df_2_light[variable].mean())
    print(df_3_light[variable].mean())
    plt.figure(figsize=(10,6), dpi=600)
    algoritmos = ['ATF', 'ACAB', 'ACAB Realista']
    promedio = [df_2_light[variable].mean(), df_3_light[variable].mean(),df_1_light[variable].mean()]
    plt.bar(algoritmos,promedio)
    if variable == 'emissions_history':
        plt.title('Promedio de emisiones de CO2 sobre tiempo simulado')
        plt.ylabel('Emisiones de CO2 en g/s')
    plt.xlabel('Algoritmos')
    plt.ylabel('Espera en segundos total')
    plt.show()
```

```python
def generate_boxplots(data_1, data_2, data_3, variable):

    # Load the dataset into a Pandas DataFrame
    df_1_light = pd.read_csv(data_1)
    variable = variable

    # Rename the column to remove an additional space
    #df = df.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})

    # Extract the date and close price columns
    step_history = df_1_light['step_history']
    variable_history = df_1_light[variable]

    # Create a line plot
    plt.plot(step_history/20, variable_history)
    plt.title('Emisiones por tiempo trfico liviano algoritmo 1')
    plt.xlabel('Segundos de simulacin')
    plt.ylabel('Emisiones de CO2 en g/s')

    df_2_light = pd.read_csv(data_2)

    # Rename the column to remove an additional space
    #df = df.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})

    # Extract the date and close price columns
    step_history = df_2_light['step_history']
    variable_history = df_2_light[variable]
    # Create a line plot
    plt.plot(step_history/20, variable_history)
    plt.title('Emisiones por tiempo trfico liviano algoritmo 2')
    plt.xlabel('Segundos de simulacin')
    plt.ylabel('Emisiones de CO2 en g/s')

    df_3_light = pd.read_csv(data_3)

    # Extract the date and close price columns
    step_history = df_3_light['step_history']
    variable_history = df_3_light[variable]
    # Create a line plot
    plt.plot(step_history/20, variable_history)
    plt.title('Emisiones por tiempo trfico liviano algoritmo 3')
    plt.xlabel('Segundos de simulacin')
    plt.ylabel('Emisiones de CO2 en g/s')
    plt.show

    print(df_1_light[variable].mean())
    print(df_2_light[variable].mean())
    print(df_3_light[variable].mean())
    #plt.figure()
    #algoritmos = ['ATF', 'AACA', 'ACSG']
    #promedio = [df_2_light[variable].mean(),df_1_light[variable].mean(), df_3_light[variable].mean()]
    #plt.bar(algoritmos,promedio)
    #plt.title('Promedio de tiempo acumulado de espera')
    #plt.xlabel('Algoritmos')
    #plt.ylabel('Espera en segundos total')
    #plt.show()

    df = pd.DataFrame({
        'Algoritmo AACA': df_1_light[variable],
        'Algoritmo ATF': df_2_light[variable],
        'Algoritmo ACAB': df_3_light[variable]
    })

    print(df)
    plt.figure()
    boxplot = df.boxplot(column=['Algoritmo AACA', 'Algoritmo ATF', 'Algoritmo ACAB'])
    plt.title('Boxplot de vehicle_history para los tres algoritmos')
    plt.ylabel('vehicle_history')
    plt.show()
```

```python
def generate_bar(data_1_L,data_1_M,data_1_H, data_2_L,data_2_M,data_2_H, data_3_L,data_3_M,data_3_H, variable):
    variable = variable
    # Load the dataset into a Pandas DataFrame
    df_1_light = pd.read_csv(data_1_L)
    df_2_light = pd.read_csv(data_2_L)
    df_3_light = pd.read_csv(data_3_L)

    df_1_med = pd.read_csv(data_1_M)
    df_2_med = pd.read_csv(data_2_M)
    df_3_med = pd.read_csv(data_3_M)

    df_1_heavy = pd.read_csv(data_1_H)
    df_2_heavy = pd.read_csv(data_2_H)
    df_3_heavy = pd.read_csv(data_3_H)


    step_history1_L = df_1_light['step_history']
    step_history2_L = df_2_light['step_history']
    step_history3_L = df_3_light['step_history']

    variable_history1L = df_1_light[variable]
    variable_history2L = df_2_light[variable]
    variable_history3L = df_3_light[variable]

    step_history1_M = df_1_med['step_history']
    step_history2_M = df_2_med['step_history']
    step_history3_M = df_3_med['step_history']

    variable_history1M = df_1_med[variable]
    variable_history2M = df_2_med[variable]
    variable_history3M = df_3_med[variable]

    step_history1_H = df_1_heavy['step_history']
    step_history2_H = df_2_heavy['step_history']
    step_history3_H = df_3_heavy['step_history']

    variable_history1H = df_1_heavy[variable]
    variable_history2H = df_2_heavy[variable]
    variable_history3H = df_3_heavy[variable]


    algoritmos = ['ATF','ACAB','ACAB Realista']
    
    promedios_L = [df_1_light[variable].mean(),df_2_light[variable].mean(),df_3_light[variable].mean()]
    promedios_M = [df_1_med[variable].mean(),df_2_med[variable].mean(),df_3_med[variable].mean()]
    promedios_H = [df_1_heavy[variable].mean(),df_2_heavy[variable].mean(),df_3_heavy[variable].mean()] 
    
    if variable == 'waiting_history':
        promedios_L = [df_1_light[variable].mean()/20,df_2_light[variable].mean()/20,df_3_light[variable].mean()/20]
        promedios_M = [df_1_med[variable].mean()/20,df_2_med[variable].mean()/20,df_3_med[variable].mean()/20]
        promedios_H = [df_1_heavy[variable].mean()/20,df_2_heavy[variable].mean()/20,df_3_heavy[variable].mean()/20]
    print("Mejora frente al algoritmo ATF")
    print(((df_3_heavy[variable].mean()))/df_1_heavy[variable].mean())
    print(((df_3_med[variable].mean()))/df_1_med[variable].mean())
    print(((df_3_light[variable].mean()))/df_1_light[variable].mean())
    print("ACAB vs ACAB realista")
    print(((df_3_heavy[variable].mean()))/df_2_heavy[variable].mean())
    print(((df_3_med[variable].mean()))/df_2_med[variable].mean())
    print(((df_3_light[variable].mean()))/df_2_light[variable].mean())    
    #print(((df_1_heavy[variable].mean()+df_3_heavy[variable].mean())/2)/df_2_heavy[variable].mean())
    x = np.arange(len(algoritmos))
    width = 0.2
    plt.figure(figsize=(10,6), dpi=600)
    plt.bar(x-width/2-width/2, promedios_L,width,label='Densidad trafico baja')
    plt.bar(x, promedios_M,width,label='Densidad trafico media')
    plt.bar(x+width/2+width/2, promedios_H,width,label='Densidad trafico alta')
    plt.xticks(x, algoritmos)
    plt.legend()
    if variable == 'emissions_history':
        plt.title('Promedio de emisiones de CO2 sobre tiempo simulado')
        plt.ylabel('Emisiones de CO2 en g/s')
        plt.ylim(0, 550)  # Set upper limit to 520
    
    if variable == 'waiting_history':
        plt.title('Promedio de tiempo acumulado de espera')
        plt.ylabel('Espera en segundos total')

    if variable == 'queue_history':
        plt.title('Promedio de cantidad de autos en cola')
        plt.ylabel('Cantidad de autos en espera total')

    if variable == 'vehicle_history':
        plt.title('Promedio de cantidad de autos activos en simulacin')
        plt.ylabel('Cantidad de autos activos')
    plt.xlabel('Algoritmos')
    plt.grid()
    plt.show()

generate_bar("emissions_dataDF2_light.csv","emissions_dataDF2_med.csv","emissions_dataDF2_heavy.csv","emissions_dataDF3_light.csv","emissions_dataDF3_med.csv","emissions_dataDF3_heavy.csv","emissions_dataDF4_light.csv","emissions_dataDF4_med.csv","emissions_dataDF4_heavy.csv",'emissions_history')
generate_bar("Queue_HistoryDF2_light.csv","Queue_HistoryDF2_med.csv","Queue_HistoryDF2_heavy.csv","Queue_HistoryDF3_light.csv","Queue_HistoryDF3_med.csv","Queue_HistoryDF3_heavy.csv","Queue_HistoryDF4_light.csv","Queue_HistoryDF4_med.csv","Queue_HistoryDF4_heavy.csv",'queue_history')
generate_bar("vehicle_dataDF2_light.csv","vehicle_dataDF2_med.csv","vehicle_dataDF2_heavy.csv","vehicle_dataDF3_light.csv","vehicle_dataDF3_med.csv","vehicle_dataDF3_heavy.csv","vehicle_dataDF4_light.csv","vehicle_dataDF4_med.csv","vehicle_dataDF4_heavy.csv",'vehicle_history')
generate_bar("waiting_dataDF2_light.csv","waiting_dataDF2_med.csv","waiting_dataDF2_heavy.csv","waiting_dataDF3_light.csv","waiting_dataDF3_med.csv","waiting_dataDF3_heavy.csv","waiting_dataDF4_light.csv","waiting_dataDF4_med.csv","waiting_dataDF4_heavy.csv",'waiting_history')
```
